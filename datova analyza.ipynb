## Jaké kombinace typů jsou vlastně možné?

### Řešení problému rovnocených typů

```python
data.head()
```

```python
data[(data["type 1"] == "Water") & (data["type 2"] == "Ice")]
```

```python
data[(data["type 2"] == "Water") & (data["type 1"] == "Ice")]
```

```python
data[["type 1", "type 2"]].values.tolist()
```

```python
data["type 2"] = data["type 2"].fillna("")
```

```python
data["types"] = data[["type 1", "type 2"]].values.tolist()
```

```python
data["types"] = data["types"].apply(lambda x: " ".join(sorted(x)).strip())
```

```python
data.groupby('types').size().sort_values(ascending=False)[:20]
```

## Úkol č. 3 - skupiny pokémonů a jejich statistiky

## A zase ten Pikachu … Jak je na tom mezi svými?

```python
pikachu_rank = data[(data['type 1'] == 'Electric') | (data['type 2'] == 'Electric')].sort_values(by='total')
pikachu_rank
```

```python
pikachu_rank = data[data['types'].str.contains('Electric')].sort_values(by='total')
pikachu_rank
```

```python
pikachu_rank.reset_index(inplace=True)
```

```python
pikachu_rank
```

```python
pikachu_rank['color'] = 'Grey'
pikachu_rank.loc[pikachu_rank['name'] == 'pikachu', 'color'] = 'Red'
pikachu_rank.plot.scatter(x='generation', y='total', color=pikachu_rank['color'])
```
